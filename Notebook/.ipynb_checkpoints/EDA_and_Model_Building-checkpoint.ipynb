#### for now as we performing clustering, so further data cleaning is not required. That is the main goal of clustering to capture the outliers and make them separate

```python
# performing k-means clustering on cc2 data
# importing `kMeans`
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# scaling the data
cc_scaled = scaler.fit_transform(cc)
cc_scaled
```

```python
import joblib
joblib.dump(scaler,"/home/ganesh/Github/PURCHASING_CAPABILITIES/Saved_models/Scaler.pkl")
```

```python
# checking the shape
cc_scaled.shape
```

```python
# creating a list to store inertia
inertia = [] 
# running for loop on different `k` vlue
for i in range(1,11):
    kmeans = KMeans(
                    n_clusters=i, init="k-means++",
                    n_init=10,
                    tol=1e-04, random_state=42
                    )
    kmeans.fit(cc_scaled) # passing the scaled value
    inertia.append(kmeans.inertia_)

    
# showing the inertia
inertia
```

```python
# plotting the inertia vs k
plt.figure(figsize=(12,6))
plt.plot(range(1,11), inertia)
plt.title("Elbow Method")
plt.xlabel("Num of K")
plt.ylabel("Inertia");
```

### For now consider `K = 3` , but it is more likely `2`

```python
kmeans = KMeans(n_clusters=3,
               init="k-means++", # default value
               tol = 1e-04, # default value
               random_state=32)

# fitting with our data X_scld
kmeans.fit(cc_scaled) # passing the scaled data
```

```python
# chekcing the labels
kmeans.labels_
```

```python
import joblib
joblib.dump(kmeans,'/home/ganesh/Github/PURCHASING_CAPABILITIES/Saved_models/cluster_model.pkl')
```

```python
# checking the silhouette score
from sklearn.metrics import silhouette_score
silhouette_score(cc_scaled, kmeans.labels_)
```

```python
# adding the cluster label to the dataframe
cc['cluster_labels'] = kmeans.labels_
cc.head(10)
```

```python
# creating another dataframe with scaled valued for PCA , so we can plot the data in 2d graph
scl_df = pd.DataFrame(cc_scaled, columns=cc.drop("cluster_labels", axis=1).columns)
# adding the cluster label
scl_df['cluster_labels'] = kmeans.labels_
scl_df.head(10)
```

```python
#  imorting pca
from sklearn.decomposition import PCA

# creating an istance of pca
pca = PCA(n_components=2)

data4 = scl_df.drop('cluster_labels', axis=1)

pca_data = pca.fit_transform(data4)
pca_data
```

```python
pca.explained_variance_ratio_
```

```python
pca_df = pd.DataFrame(pca_data, columns=['pca1', 'pca2'])
pca_df['labels'] = kmeans.labels_
pca_df
```

```python
# plotting all the data
plt.figure(figsize=(12,6))
sns.scatterplot(x='pca1', y='pca2', data=pca_df, hue='labels');
```

### Creating polar graph

```python
# checking the unique values of cluster labels
scl_df['cluster_labels'].value_counts()
```

```python
# groupby the label column according to all other features
polar = scl_df.groupby("cluster_labels").mean().reset_index()
polar
```

```python
melt_df = pd.melt(polar, id_vars=['cluster_labels'])
melt_df
```

```python
# importing potly 
import plotly.express as px

fig4 = px.line_polar(melt_df, r="value", theta="variable", color="cluster_labels", 
                     line_close=True,height=650,width=1100)
fig4.show()
```


### Conclusion:

From the above ploar graph we can separate 3 cluster and separate customer based on the cluster.


`0 - cluster` - these customers are having `High` Purchase capability
